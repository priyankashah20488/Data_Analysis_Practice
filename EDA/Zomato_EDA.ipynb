```python
df_country=pd.read_excel('country_code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_name
```

```python
country_val=final_df.Country.value_counts().values
```

```python
country_val
```

```python
# pie chart - top 3 countries that use zomato
plt.pie(country_val,labels=country_name)
```

```python
#top 3 countries that uses Zomato
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation:
- Zomato maximum records or trancation in India
- After that USA and United kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation:
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when rating is betweeen 4.0 to 3.4---> very good
3. when rating is betweeen 3.5 to 3.9---> good
4. when rating is betweeen 3.0 to 3.4---> average
5. when rating is betweeen 2.5 to 2.9---> average
6. when rating is betweeen 2.0 to 2.4---> poor

```python
ratings.head()
```

```python
import matplotlib 
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:
    
    1.Not rated count is very high
    2.maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
