```python
l = list()
while True:
  n = input('enter number: ')
  if n.isnumeric():
   l.append(n)
  else:
    print("Invalid input")
  m = input('do you want to add another? enter y to continue. ')
  if m != 'y':
    break
print("Your Input List is:",l)

for i in range(len(l)):
  l[i] = int(l[i])
print("converted in integer",l)
```

```python
l = list()
n = input("Enter numbers using , operator:")
num = n.split(',')
for i in num:
  if i.isnumeric():
    l.append(i)
  else:
    print("Invalid Input Only Numbers are allowed.")
    break
print(l)  
for s in range(len(l)):
  l[s]=int(l[s])
print("converted into integer:",l)      
```
