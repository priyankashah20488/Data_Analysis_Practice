```python
#alternate of formate function
def string_formate(str_source,*values):
  count_placefolder =  str_source.count("__")
  str_source = str_source.split()
  c = 0
  if count_placefolder == len(values):
    for s in str_source:
      if '__' in s:
        placeholder_index = str_source.index(s)
        str_source[placeholder_index] = s.replace('__',values[c])
        c+=1
    str_source = " ".join(str_source)
    return str_source
  else:
    print("invalid input")

```

```python
# calling sample 2 - str_format("hello _name_! welcome to _place_", name='kush', place='ahmedabad')

def string_format(str_source,**args):
  count_placeholder = str_source.count('_')
  str_source = str_source.split()
  len_placeholder = count_placeholder//2
  #print(len_placeholder)
  if len_placeholder == len(args):
    for i in str_source:
      if i.startswith("_") and i.endswith('_'):
        a = i[1:len(i)-1]
        if a in args.keys():
          placeholder_index = str_source.index(i)
          str_source[placeholder_index] = i.replace(i,args[a])

      if i.startswith("_"):
        a = i[1:len(i)-2]
        if a in args.keys():
          placeholder_index = str_source.index(i)
          str_source[placeholder_index] = i.replace(i[0:len(i)-1],args[a])

    str_source = " ".join(str_source)

    return str_source
  else:
    print("invalid input")
```

```python
string_format("hello _name_! welcome to _place_", name='kush', place='ahmedabad')
```
