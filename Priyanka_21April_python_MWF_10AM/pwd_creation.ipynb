```python
# how to generate password 

import random

len_pwd = 8
lower_case = ''
for d in range(1,3):
    for c in range(97, 123):
       lower_case+=chr(c)
       #print(chr(c))
    
upper_case = ''
for s in range(64,91):
    upper_case+=chr(s)
    
Num = ''
for n in range(48,59):
    Num+=chr(n)

sp_chr=''
for i in range(33, 96):
  if not chr(i).isalnum() and '\\' and '/' and '.' and ',' and ':' and '>' and '<' and '(' and ')' and ']' and '[':
    sp_chr+=chr(i)

combined = lower_case+upper_case+Num+sp_chr

sp_lower = random.choice(lower_case) 
sp_upper = random.choice(upper_case)
sp_num = random.choice(Num)
sp_char = random.choice(sp_chr)

temp_pwd = sp_lower + sp_upper + sp_num + sp_char

for i in range(len_pwd-4):
  pwd = temp_pwd + random.choice(combined)

New_pwd = pwd.split()
random.shuffle(New_pwd)
print(New_pwd)


l = " ".join(New_pwd)
print(l)
```
