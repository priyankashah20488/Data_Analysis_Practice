```python
# create class
class Calculator:
  def __init__(self):
    self.result = None

  def set_add(self,x,y,*num):
    result=x+y
    for i in num:
      result+=i
      self.result=result
    return result

  def set_subtract(self,x,y,*num):
    result = x-y
    for i in num:
      result-=i
      self.result = result
    return result

  def set_multiplication(self,x,y,*num):
    result = x*y
    for i in num:
      result*=i
      self.result = result
    return result

  def set_div(self,x,y):
    result = x/y
    self.result = result
    return result
```

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):
  def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)

x = Student("Mike", "Olsen")
x.printname()
```

```python
class Person:
  def __init__(self):
    self.name = None

  def set_name(self,name):
    self.name = name

  def get_name(self):
    print(self.name)

class Manager(Person):
  def info(self):
    self.dep_name = 'IT Dep'

  def get_info(self):
    print(self.name,self.dep_name)
```

```python
obj = Manager()
```

```python
obj.set_name('Priyanka')
```

```python
obj.get_name()
```

```python
obj.info()
```

```python
obj.get_info()
```
