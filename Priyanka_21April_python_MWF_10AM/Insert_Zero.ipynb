```python
numbers = [2,3,4,5,6,7]
for i in range(1,11):
    if i < 10:
   # i = f'{i:02}'
     i1 = str(i).zfill(3)
     i1 = int()

    for n in numbers:
     a = i1 * n
     if a < 10:
      # a = f'{a:02}'
      a1 = str(a).zfill(3)
      a1 = int() 

print(f"{n} * {i1} = {a1}", end="\t|\t")      
print()
```
