```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = [1,2,3,4,5,6,500]
mean = np.mean(dataset)
print(mean)
std = np.std(dataset)
print(std)
thrasehold = 2
outliers = []
for i in dataset:
  z = (i-mean)/std
  if np.abs(z) > thrasehold:
    outliers.append(i)
print("outliers is:",outliers)
```

##IQR
1. sorrt the data
2. calculate Q1(25%) and Q3(75%)
3. IQR(Q3-Q1)
4. Find the Lower Fence (q1-1.5(iqr))
5. Find the upper fence (q3+1.5(iqr))


```python
##sort
dataset=sorted(dataset)
```

```python
dataset
```

```python
q1,q3=np.percentile(dataset,[25,75])
print(q1,q3)
```

```python
iqr=q3-q1
print(iqr)
```

```python
##find the lower fence and upper fence
lower_fence=q1-(1.5*iqr)
upper_fence=q3+(1.5*iqr)
print(lower_fence,upper_fence)
```

```python
outliers=[]
for i in dataset:
  if (i < lower_fence) or (i > upper_fence):
    outliers.append(i)
#print(outliers)
```
