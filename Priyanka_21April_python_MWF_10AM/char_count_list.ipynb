```python
string = "hello welcome to logicrays"
l1 = list()
l2 = list()
for i in string:
  if i not in l1:
    l1.append(i)
for l in l1:
  c = 0
  for s in string:
    if s == l:
     c += 1    
     l2.append((l,c))
print(l2)
```
