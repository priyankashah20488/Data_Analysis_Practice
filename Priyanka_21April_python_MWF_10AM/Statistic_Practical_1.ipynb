Measure of Dispersion

1.Variance
2.Standard deviation

```python
statistics.variance(ages)

#differnce in np.var and statistics.var beause in statistics it use sample variance formula that is divide by n-1
```

```python
#differnce in np.var and statistics.var beause in statistics it use population variance that is divede by N
```

```python
def variance(data):
  n=len(data)

  ##mean of data
  mean_data=sum(data)/n

  ##variance
  deviation=[(i-mean_data)**2 for i in data]
  variance=sum(deviation)/(n-1)
  return variance
```

```python
variance(ages)
```

```python
# if dof =0 than variance of population data and dof = 1 than variance of sample data
def variance(data,dof=0):
  n=len(data)
   ##mean of data
  mean_data=sum(data)/n

  ##variance
  deviation=[(i-mean_data)**2 for i in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
variance(ages,1)
```

```python
variance(ages,0)
```

```python
## in statistic find varionce of population use pvariance()
statistics.pvariance(ages)
```

```python
import math
```

```python
#find standard deviation using math
math.sqrt(statistics.pvariance(ages))
```

Check whether Distribution is Normal Distribution

```python
# if u want to check whether feature is gaussian or normal distributed
# Q-Q plot

import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot=pylab)
  plt.show()
```
